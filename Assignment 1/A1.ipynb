Our dataset contains weekly retail prices of three grades of gasoline, diesel, auto propane and compressed natural gas in 10 Ontario markets.
```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('energy-price.csv')
df.head(10)
```

## Question 1

Compare the mean prices of Regular Unleaded Gasoline among different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") in 2023, and plot a corresponding bar chart. Please ensure that the bars are sorted by the price amounts.

```python


cities = ['Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's"]

# Filter for 2023 and Regular Unleaded Gasoline
new_df = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Date'].str.contains('2023'))]

# Calculate mean prices
mean_prices = {city: new_df[city].mean() for city in cities}

# Sort cities by mean price
sorted_cities = sorted(mean_prices, key=mean_prices.get)

# Plot
plt.bar(sorted_cities, [mean_prices[city] for city in sorted_cities])
plt.xticks(rotation=90)
plt.ylabel('Mean Price')
plt.title('Mean Prices of Regular Unleaded Gasoline in 2023')
plt.show()
```

## Question 2
Plot the corresponding dot plot for the same information presented in Question 1.

```python
plt.figure(figsize=(10, 8))
plt.plot([mean_prices[city] for city in sorted_cities], sorted_cities, 'o')
plt.xlabel('Mean Price')
plt.title('Mean Prices of Regular Unleaded Gasoline in 2023')
plt.show()
```

## Question 3

Plot a histogram of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023, using 15 bins.

```python
df_toronto_west = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Date'].str.contains('2023')) & (df['Toronto West/Ouest'].notna())]

plt.hist(df_toronto_west['Toronto West/Ouest'], bins=15)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Regular Unleaded Gasoline Prices in Toronto West/Ouest for 2023')
plt.show()
```

## Question 4

Plot the normalized ECDF (Empirical Cumulative Distribution Function) of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023.

```python
prices = df_toronto_west['Toronto West/Ouest'].sort_values()

cumulative_probabilities = np.arange(1, len(prices) + 1) / len(prices)

plt.plot(prices, cumulative_probabilities, marker='.', linestyle='none')
plt.xlabel('Price')
plt.ylabel('ECDF')
plt.title('ECDF of Regular Unleaded Gasoline Prices in Toronto West/Ouest for 2023')
plt.show()
```

## Question 5

Plot a boxplot to display the distribution of prices across different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") from 2021 onwards (including 2021).


```python
df_2021_onwards = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Date'].str.contains('202[1-9]'))]

data_to_plot = [df_2021_onwards[city].dropna() for city in cities]

# Plot
plt.figure(figsize=(10, 8))
plt.boxplot(data_to_plot, vert=False)
plt.yticks(np.arange(1, len(cities) + 1), cities)
plt.xlabel('Price')
plt.title('Distribution of Regular Unleaded Gasoline Prices from 2021 Onwards')
plt.show()
```
