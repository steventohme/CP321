```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('energy-price.csv')
df.head(10)
```

## Question 1

Compare the mean prices of Regular Unleaded Gasoline among different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") in 2023, and plot a corresponding bar chart. Please ensure that the bars are sorted by the price amounts.

```python


cities = ['Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's"]

# Filter for 2023 and Regular Unleaded Gasoline
new_df = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Date'].str.contains('2023'))]

# Calculate mean prices
mean_prices = {city: new_df[city].mean() for city in cities}

# Sort cities by mean price
sorted_cities = sorted(mean_prices, key=mean_prices.get)

# Plot
plt.bar(sorted_cities, [mean_prices[city] for city in sorted_cities])
plt.xticks(rotation=90)
plt.ylabel('Mean Price')
plt.title('Mean Prices of Regular Unleaded Gasoline in 2023')
plt.show()
```

## Question 2
Plot the corresponding dot plot for the same information presented in Question 1.

```python
plt.figure(figsize=(10, 8))
plt.plot([mean_prices[city] for city in sorted_cities], sorted_cities, 'o')
plt.xlabel('Mean Price')
plt.title('Mean Prices of Regular Unleaded Gasoline in 2023')
plt.show()
```
